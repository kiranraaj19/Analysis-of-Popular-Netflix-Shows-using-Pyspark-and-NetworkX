```python
# For Creating Networks

import networkx as nx
import pandas as pd
import plotly.graph_objects as go

# For drawing Graphs
import matplotlib.pyplot as plt

# For generating random color
import random
```

```python
df = pd.read_csv("./netflix_db.csv")
## add new features in the dataset
df["date_added"] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

df['season_count'] = df.apply(lambda x : x['duration'].split(" ")[0] if "Season" in x['duration'] else "", axis = 1)
df['duration'] = df.apply(lambda x : x['duration'].split(" ")[0] if "Season" not in x['duration'] else "", axis = 1)
```

```python
## EDGES

edges = []

df = df.head(2000)

for index, row in df.iterrows():
    temp = row['listed_in'].split(',')
    for i in range(len(temp)):
        for j in range(i+1,len(temp)):
            edges.append([temp[i],temp[j]])
            
print(len(edges))
```

```python
# NAME -> ID
cat2 = {}
cnt = 1
for index, row in df.iterrows():
    for x in row['listed_in'].split(','):
        if x in cat2:
            pass
        else:
            cat2[x]=cnt
            cnt+=1
            

node_A = []
node_B = []

for x in edges:
    node_A.append(cat2[x[0]])
    node_B.append(cat2[x[1]])

print(node_A)
print(node_B)

di = {"Source":node_A,"Target":node_B}

df = pd.DataFrame(di)

df
```

```python
df.to_csv('edges.csv',index=False) 
```
